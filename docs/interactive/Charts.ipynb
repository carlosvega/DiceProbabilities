```python
#init_notebook_mode(connected=True)
%matplotlib inline
import matplotlib, plotly, fileinput
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
from math import factorial, floor
from collections import Counter
import seaborn as sns

from plotly.offline import plot, iplot, init_notebook_mode
import plotly.graph_objs as go
init_notebook_mode(connected=True)
```

```python
def comb_sans_repeat(n, k):
    return factorial(n) / float(factorial(k)*factorial(n-k))

def dice_prob(p, n, s):
    k_max = floor((p-n)/float(s))+1
    acc = 0
    for k in range(k_max):
        val1 = (-1)**k;
        val2 = comb_sans_repeat(n, k)
        val3 = comb_sans_repeat(p-(s*k)-1, n-1)
        acc += (val1 * val2 * val3)
    return acc / float(s**n)

def dice_prob_distr(d, t, s):
    max_number  = s*(d*t);
    min_number  = (d*t);
    possible_results = np.linspace(min_number, max_number, max_number-min_number+1);

    probabilities = []
    for i in range(len(possible_results)):
        probabilities.append(dice_prob(possible_results[i], t*d, s))
    return probabilities, possible_results
```

```python
probs, values = dice_prob_distr(3, 1, 6)
bar = go.Bar(x=values, y=probs, hoverinfo='label+percent', textfont=dict(size=20), name='None') 
layout = dict(title='Select the number of dice', showlegend=True)
fig = dict(data=[bar], layout=layout)

iplot(fig, filename='styled_pie_chart')
```

```python
dices = range(1, 11)
colours = sns.color_palette("pastel", len(dices)).as_hex()

empty_trace = go.Bar(x=[], y=[],
               hoverinfo='label+percent', 
               textfont=dict(size=20), name='None')

traces = [empty_trace]
for i, n_dices in enumerate(dices):
    probs, values = dice_prob_distr(n_dices, 1, 6)
    bar = go.Bar(x=values, y=probs, hoverinfo='label+percent', textfont=dict(size=20), name='{} dices'.format(n_dices), marker=dict(
        color=colours[i])) 
    traces.append(bar)
```

```python
#button list

keys = ['Select a number of dices']
keys += ['{} dice{}'.format(n_dices, 's' if n_dices > 1 else '') for n_dices in dices]

button_list = []
for i, n_dices in enumerate(keys):
    bools = [False]*len(keys)
    if i == 0:
        button = dict(label = str(n_dices),
          method = 'update',
          args = [{'visible': bools},
          {'title': 'Select the number of dices'}])
    else:
        bools[i] = True
        button = dict(label = str(n_dices),
          method = 'update',
          args = [{'visible': bools},
          {'title': 'Probabilities for obtaining a given total using {} dices with 6 sides'.format(n_dices)}])
    #add
    button_list.append(button)
```

```python
updatemenus = list([
    dict(active=0,
         direction = 'down',
         pad = {'r': 10, 't': 10},
        showactive = True,
        x = 0,
        xanchor = 'left',
        y = 1,
        yanchor = 'bottom' ,
         buttons=button_list,
    )
])

layout = dict(title='Select the number of dices', showlegend=True, updatemenus=updatemenus, xaxis=dict(title='Value'), yaxis=dict(title='Probability'))

fig = dict(data=traces, layout=layout)

iplot(fig, filename='styled_bar_chart')
plot(fig, filename='styled_bar_chart')
```
